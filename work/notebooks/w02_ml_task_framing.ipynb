## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*



**Task Type:** Classification

This is a classification problem because the goal is to predict whether a webpage is declining or not based on its historical SEO and content features. The model assigns each page to one of two classes:

- 1 = Declining page
- 0 = Not declining page

The prediction helps identify pages that may need content updates or optimization.

```python
import pandas as pd
from pathlib import Path

# Automatically locate the project root
current = Path.cwd()

# Search upward until we find the "data" folder
while not (current / "data").exists():
    if current.parent == current:
        raise FileNotFoundError("Could not find the project's data folder.")
    current = current.parent

# Dataset path
data_path = current / "data" / "raw" / "content_refresh_anonymized.csv"

print("Dataset Path:", data_path)

# Load dataset
df = pd.read_csv(data_path)

# Create target label
df["is_declining_label"] = (
    df["trend_direction"]
      .str.lower()
      .eq("down")
      .astype(int)
)

# Display basic information
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nTarget Distribution:")
print(df["is_declining_label"].value_counts())

print("\nSample Data:")
display(df[["trend_direction", "is_declining_label"]].head())
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



**Target:** `is_declining_label`

The model predicts whether a webpage is declining or not.

- **1** = Declining page
- **0** = Not declining page

The target is not directly provided in the dataset. It is created using the `trend_direction` column.

If `trend_direction` is **"down"**, the page is labeled as **1 (Declining)**. Otherwise, it is labeled as **0 (Not Declining)**.

This is a **defined rule (proxy label)** rather than a directly observed outcome.

```python
# Create the target label
df["is_declining_label"] = (
    df["trend_direction"]
      .str.lower()
      .eq("down")
      .astype(int)
)

# Show the target column
df[["trend_direction", "is_declining_label"]].head(10)
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*



**Success Metric:** Precision@50

Precision@50 measures how many of the top 50 webpages predicted as declining are actually declining.

A higher Precision@50 means the model is better at identifying the pages that truly need attention.

For this task, a good model should achieve a higher Precision@50 than the hand-written baseline rule, ensuring that content teams spend their time on the most important pages first.

```python
print("Success Metric: Precision@50")
print("Goal: Maximize the number of truly declining pages in the top 50 predictions.")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



A fixed rule uses simple conditions, such as:

- If `content_age_days > 365`, then update the page.
- If `impressions_90d < 500`, then ignore the page.

These rules consider only one or two features and cannot capture the complex relationships between multiple factors.

Machine Learning can learn patterns from several features at the same time, including:
- Content age
- Days since last update
- Impressions
- Average position
- CTR
- Word count

The model identifies combinations of these features to predict whether a webpage is declining more accurately than a single hand-written rule. This helps content teams prioritize pages that are most likely to benefit from updates.
